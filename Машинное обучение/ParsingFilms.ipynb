```python
import requests
from bs4 import BeautifulSoup as bs

result_list = {
    'title': [], 'title': [], 'year': [],
    'description': [], 'rating': [], 'duration': []
}

pagenum = 1
for _ in range(10):
    url = f'https://ru.kinorium.com/collections/imdb/233/{pagenum}/' 
    page = requests.get(url)
    soup = bs(page.text, 'html.parser')

    titles = soup.find_all('a', class_='titlefilm')  

    if not titles:
        print( )
        break  

    for i in titles:
        film_url = 'https://ru.kinorium.com' + i.get('href') 
        film_page = requests.get(film_url)
        film_soup = bs(film_page.text, 'html.parser')

      
        title_ru = film_soup.find('h1', class_="film")
        title_eng = film_soup.find('div', itemprop="alternativeHeadline")

        
        release_date_world = film_soup.find('td', text="Премьера в мире:")
        if release_date_world:
            release_date_world = release_date_world.find_next_sibling('td').text.strip()

        
        description = film_soup.find('div', itemprop="description")
        if description:
            description = description.text.strip()

      
        rating_tag = film_soup.find('meta', itemprop="ratingValue")
        recommend_count = rating_tag.get('content') if rating_tag else 'N/A'

 
        duration_row = film_soup.find('td', string="Продолжительность:")
        duration = duration_row.find_next_sibling('td').text.strip() if duration_row else 'N/A'

        if title_ru:  
            result_list['title'].append(title_ru.text.strip())
            result_list['english_title'].append(title_eng.text.strip() if title_eng else 'N/A')
            result_list['year'].append(release_date_world)
            result_list['description'].append(description)
            result_list['rating'].append(recommend_count)
            result_list['duration'].append(duration)

    pagenum += 1  

print(result_list)  
```

```python
result_list
```

```python
file_name = 'films3.csv'
df = pd.DataFrame(data=result_list)
df.to_csv(file_name)
```
