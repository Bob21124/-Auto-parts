```python
def extract_text_from_pdf(pdf):
    doc = fitz.open(pdf)
    text = []
    for current_page in range(len(doc)):
        page = doc.load_page(current_page)
        page_text = page.get_text("text")
        text.append(page_text)
    return text
```

```python
def pars_pdf(text):
    text2 = text.split('\n')
    print(text2)
    if '–†–µ–π—Ç–∏–Ω–≥' in text2:
        num = text2.index('–†–µ–π—Ç–∏–Ω–≥')
        raiting = text2[num-1]
        name = text2[num+1]
        activity = text2[num+2]
        date_publish = text2[num+3]
        description = text2[num+4]
        print(num, raiting, name, activity, date_publish, description)
        
    else:
        raiting = 0
        name = ""
        activity = ""
        description = ""
        if '–ü—Ä–æ—Å—Ç–æ–π' in text2:
            num2 = text2.index('–ü—Ä–æ—Å—Ç–æ–π')
            date_publish = text2[num2-1]
        else:
            date_publish = ""
        print(num, raiting, name, activity, date_publish, description)
    
    return  name, raiting, activity, date_publish, description
```

```python

Brak = []
NameCompany = []
Description = []
Raiting = []
DataPublish = []
Activity = []
TextArticle = []
for pdf in all_pdf:
    try:  
        x = extract_text_from_pdf(pdf)
        print(pdf.split('\\')[-1])
        N, R, A, D, Desc = pars_pdf(x[0])
        T = ''.join(x).replace('\n',' ')
        
        
        NameCompany.append(N)
        Raiting.append(R)
        DataPublish.append(D)
        Activity.append(A)
        Description.append(Desc)
        TextArticle.append(T)
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞–±–æ—Ç–∫–µ —Ñ–∞–π–ª–∞ {pdf}: {e}")
        Brak.append(pdf)
```
